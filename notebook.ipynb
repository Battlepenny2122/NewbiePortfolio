## Analyzing Industry Carbon Emissions

For this dataset, I was asked to find which industries were emitting the most carbon, and it's total carbon footprint.
This was done unguided, and demonstrates simple aggregation and filtering.

```python
SELECT
industry_group,
count(*) AS count_industry,
ROUND(SUM(carbon_footprint_pcf),1) AS total_industry_footprint
FROM product_emissions
WHERE year = 2017
GROUP BY industry_group
ORDER BY total_industry_footprint DESC
```

## User's Third Transaction [Uber SQL Interview Question]

For this dataset, I was required to find the third transaction for each user of uber. This query demonstrates my ability to use CTEs and Temp tables, as well as some simple windows functions.

Available at: https://datalemur.com/questions/sql-third-transaction

```python
WITH cte AS 
(Select user_id, spend, transaction_date, RANK() OVER (PARTITION BY user_id ORDER BY transaction_date) AS rank
FROM transactions)

SELECT 
user_id, spend, transaction_date
FROM cte
WHERE rank=3;
```

## The report [HackerRank intermediate question]

This question required a case function, inequalities, and a join to determine a specific report for a teacher's students.
Available here: https://www.hackerrank.com/challenges/the-report/

```python
SELECT 
(CASE WHEN Grade >7 THEN a.name ELSE a.name=NULL END) AS Name,
b.grade,
a.marks
FROM students a
INNER JOIN grades b
ON a.marks <= b.Max_Mark AND a.marks >= b.Min_Mark
ORDER BY b.grade DESC, Name
```
